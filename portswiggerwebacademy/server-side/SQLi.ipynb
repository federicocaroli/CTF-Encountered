````python
# Python version of the payload function of the JS file

def payload(data):
    """
    Costruisce e restituisce una stringa XML a partire da un dizionario di coppie chiave-valore.

    :param data: dict contenente le coppie chiave-valore da inserire nel payload XML
    :return: stringa XML formattata
    """
    # Intestazione XML
    xml = '<?xml version="1.0" encoding="UTF-8"?>'
    xml += '<stockCheck>'

    # Iterazione sulle coppie chiave-valore
    for key, value in data.items():
        xml += f'<{key}>{value}</{key}>'

    # Chiusura dell'elemento radice
    xml += '</stockCheck>'
    return xml

# Python script to convert letters in hex ref
def to_hex_references(text: str, uppercase: bool = False) -> str:
    """
    Convert each character in `text` into an XML hexadecimal numeric reference.
    
    Args:
        text:      Input string to encode.
        uppercase: If True, uses uppercase hex digits (&#x41;  &#x41; vs &#x61;).
    
    Returns:
        A new string where every character is replaced by '&#xHH;'.
    """
    
    res = ''
    for c in text:
        res += f'&#{hex(ord(c))[1:]};'

    return res
````
